```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load cleaned data
df = pd.read_csv("processed_data.csv")
```

```python
# Correlation heatmap
plt.figure(figsize=(14, 8))
# Filter only numeric columns
numeric_df = df.select_dtypes(include='number')

# Plot correlation
plt.figure(figsize=(14, 8))
sns.heatmap(numeric_df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
# Department-wise performance
df_dept = pd.read_excel("INX_Future_Inc_Employee_Performance_CDS_Project2_Data_V1.8.xls")
avg_rating = df_dept.groupby('EmpDepartment')['PerformanceRating'].mean().sort_values()
avg_rating.plot(kind='bar', color='skyblue')
plt.title("Avg Performance by Department")
plt.ylabel("Rating (1-4)")
plt.xticks(rotation=45)
plt.show()
```
